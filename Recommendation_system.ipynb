```python
watched_by_user1=list(set(user_1.genre))
watched_by_user2 = list(set(user_2.genre))

if len(watched_by_user1) > len(watched_by_user2):
    for genre in watched_by_user1:
      if genre not in watched_by_user2:
        print("Recommendation : ",genre)
elif len(watched_by_user1)<len(watched_by_user2):
    for genre in watched_by_user2:
      if genre not in watched_by_user1:
         print("Recommendation : ",genre)
else:
    print("No recommendation")
```
